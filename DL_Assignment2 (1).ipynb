```python
import tensorflow
from tensorflow import keras
```

```python
# loading keras mnist dataset
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
# x_train is in 3D array form, but we need 2D array of x_train ,neural network need x_train in matrix form
# reshape 3D images into 2D vectors
x_train = x_train.reshape(x_train.shape[0], -1)
x_test  = x_test.reshape(x_test.shape[0], -1)

print(x_train.shape)
print(x_test.shape)
```

```python
#scaling: using min-max scaling-values scaled to [0,1]
x_train=x_train/255.0
x_test=x_test/255.0
```

```python
import numpy as np
#initalise weights and bias and learning rate
n=0.01
weights=np.random.rand(784,10)*0.01
bias=np.random.rand(1,10)*0.01
print(weights)
print(bias)
```

```python
#define softmax activation function
def softmax(z):
  z_stable= z - np.max(z) # to avoid overflow exp(1000), subtracting max(z) from all values of z don't change softmax value
  exp_z= np.exp(z_stable)
  return exp_z/np.sum(exp_z)
```

```python
#define categorical cross entropy

y_train_digit=y_train
y_test_digit=y_test


#convert o/p class to one hot encoding
def one_hot_encode(y, num_classes):
    one_hot = np.zeros((len(y), num_classes))

    for i in range(len(y)):
        one_hot[i][y[i]] = 1

    return one_hot

y_train_digit=one_hot_encode(y_train_digit,10)
y_test_digit=one_hot_encode(y_test_digit,10)


```

```python
#define loss function
import numpy as np

def categorical_loss(y, y_hat):
    return -np.sum(y * np.log(y_hat + 1e-9)) # avoid log(0)
```

```python
print(y_train)
```

```python
def Training(x_train, y_train):
    global weights, bias
    epochs = 10

    for epoch in range(epochs):
        total_loss = 0

        for i in range(x_train.shape[0]):
            x = x_train[i:i+1]
            y = y_train[i:i+1]

            # Forward
            z = x @ weights + bias
            y_hat = softmax(z)

            # Loss
            loss = categorical_loss(y, y_hat)
            total_loss += loss   # ✅ FIXED

            # Backward
            dz = y_hat - y
            dw = x.T @ dz
            db = np.sum(dz, axis=0, keepdims=True)

            # Update
            weights -= n * dw
            bias -= n * db

        print(f"Epoch {epoch+1}, Loss: {total_loss / x_train.shape[0]:.4f}")

Training(x_train, y_train)
```

```python
#getting weights and bias
print(weights)
print(bias)
```

```python
def prediction(x_test):
    # Ensure 2D shape
    if x_test.ndim == 1:
        x_test = x_test.reshape(1, -1)

    # Forward pass
    z = x_test @ weights + bias
    y_pred = softmax(z)

    # Class prediction
    return np.argmax(y_pred, axis=1)
prediction(x_test)
```

```python
#performance
def accuracy(x, y_true):
    # Predict class labels
    y_pred = prediction(x)

    # Convert one-hot → digit labels (if needed)
    if y_true.ndim == 2:
        y_true = np.argmax(y_true, axis=1)

    # Accuracy calculation
    acc = np.mean(y_pred == y_true)
    return acc
```

```python
train_acc = accuracy(x_train, y_train)
test_acc  = accuracy(x_test, y_test)

print("Train Accuracy:", train_acc)
print("Test Accuracy :", test_acc)
```
